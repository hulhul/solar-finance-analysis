```python
sfdWeather = pd.read_csv("SolarFinanceProjectWeatherData.csv")
```

```python
sfdWeather.head(10)
```

```python
sfdWeatherRedone = sfdWeather.groupby('name', as_index=False).mean()
sfdWeatherRedone
```

```python
sfd = sfd.sort_values('City')
sfd
```

```python
sfdUpdated = sfd.set_index("City").join(sfdWeatherRedone.set_index("name"))
sfdUpdated
```

```python
sfdUpdatedNew = sfdUpdated.drop(columns=["NO2      AM (ppb)", "NO2          1-hr (ppb)", "SO2         1-hr (ppb)"])
```

```python
sfdUpdatedNew.to_csv("SOLARFINANCEFINAL.csv")
```
